```python
from DeepHedging.HedgingInstruments import GBMStock, HestonStock
```

```python
gbm = GBMStock(S0=100, T=1.0, N=252, r=0.05, sigma=0.2)
gbm_paths = gbm.generate_paths(num_paths=10000)
gbm.plot(gbm_paths[:10], title="GBM Stock Price Paths")
```

```python
heston = HestonStock(S0=100, T=1.0, N=252, r=0.05, v0=0.04, kappa=2.0, theta=0.04, xi=0.1, rho=-0.7)

heston_paths, variance_paths = heston.generate_paths(num_paths=10000, return_variance=True)

heston.plot(heston_paths[:10], title="Heston Stock Price Paths")
heston.plot(variance_paths[:10], title = 'Heston Stock Variance Paths')
```
